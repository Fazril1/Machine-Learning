```python
import numpy as np # Memanggil module numpy untuk melakukan fungsi aritmatika dalam kode
import pandas as pd # Memanggil module pandas untuk melakukan fungsi pembacaaaan dataframe berbentuk tabel
from sklearn.model_selection import train_test_split # Memanggil module untuk menggunakan fungsi train_test_split untuk membagi dataset menjadi 2 yakni pengujian dan pelatihan
from sklearn.tree import DecisionTreeRegressor # Memanggil module sklearn.tree untuk menggunakan metode decision tree
from sklearn.ensemble import RandomForestRegressor # Memanggil sklearn.ensemble untuk menggunakan metode random forest
from sklearn.metrics import mean_squared_error # Memanggil module sklearn.metrics untuk menggunakan fungsi MSE
from sklearn.preprocessing import LabelEncoder # Memanggil module sklearn.preprocessing untuk menggunakan fungsi data processing
```

```python
data = pd.read_csv('data.csv') # membaca dataset yang diberikan
```

```python
#Pra-proses dataset agar kode bisa mengolah datanya untuk diolah
data['date'] = pd.to_datetime(data['date']) # Mengonversi kolom 'date' menjadi bentuk numerik
data['numeric_date'] = data['date'].astype(int) / 10**9  # Konversi ke detik

# Menghapus kolom asli 'date' karena sudah tidak diperlukan
data = data.drop('date', axis=1)

# Inisialisasi LabelEncoder untuk mengubah string menjadi bentuk numerik
label_encoder = LabelEncoder()

# # Mengonversi kolom 'street' menjadi bentuk numerik
data['encoded_street'] = label_encoder.fit_transform(data['street'])

# Loop melalui semua kolom dalam dataframe dan terapkan Label Encoding
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = label_encoder.fit_transform(data[column])

# Menghapus kolom asli 'street' karena sudah tidak diperlukan
data = data.drop('street', axis=1)
```

```python
X = data.drop('price', axis=1) # Data dalam dataset selain data target yaitu "price" akan dimasukkan ke variabel x untuk di uji dan dilatih
y = data['price'] # Data dalam dataset yaitu "price" akan dimasukkan ke variabel y untuk di uji dan dilatih
```

```python
# kode untuk membagi dataset menjadi 2 bagian yaitu tes dan pelatihan data dengan menggunakan; test size 0,3 berarti 30% dari data akan digunakan untuk pengujian dan sisanya untuk pelatihan; random state digunakan untuk membuat hasil prediksi selalu konsisten
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
decision_tree_model = DecisionTreeRegressor(random_state=42) # Memanggil metode decision tree yang disimpan dalam variabel decision_tree_model
decision_tree_model.fit(X_train, y_train) # Simpan data yang sudah dilatih di variabel decision_tree_model.fit
y_pred_decision_tree = decision_tree_model.predict(X_test) # 30% data yang dites selain kolom price sebagai parameter untuk memprediksi harga rumah disimpan dalam variabel y_pred_decision_tree
```

```python
random_forest_model = RandomForestRegressor(random_state=42) # Memanggil metode random forest yang disimpan dalam variabel random_forest_model
random_forest_model.fit(X_train, y_train) # Simpan data yang sudah dilatih di variabel random_forest_model.fit
y_pred_random_forest = random_forest_model.predict(X_test) # 30% data yang dites selain kolom price sebagai parameter untuk memprediksi harga rumah disimpan dalam variabel y_pred_random_forest
```

```python
# Mean Squared Error untuk melihat seberapa handal metode tersebut untuk memprediksi harga rumah, semakin kecil nilainya maka semakin handal metode tersebut memprediksi harga rumah dengan rumus (1/n) * Σ(yi - ŷi)^2, n = jumlah baris data yang diuji; yi = nilai harga yang sebenarnya; ŷi = nilai prediksi harga
mse_decision_tree = mean_squared_error(y_test, y_pred_decision_tree) # Simpan hasil mse ke variabel mse_decision_tree
mse_random_forest = mean_squared_error(y_test, y_pred_random_forest) # Simpan hasil mse ke variabel mse_random_forest
```

```python
print(f'MSE Decision Tree: {mse_decision_tree}') # Tampilkan nilai MSE dari metode decision tree
print(f'MSE Random Forest: {mse_random_forest}') # Tampilkan nilai MSE dari metode random forest
```
